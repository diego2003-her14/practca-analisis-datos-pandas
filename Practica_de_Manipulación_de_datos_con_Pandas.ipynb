## Verifica si la base de datos contiene datos nulos y, en caso de tenerlos, realiza el tratamiento de estos datos nulos de la manera que consideres ms coherente con la situacin.

```
datos.isnull().sum()
```

```
datos = datos.fillna(0)
```

```
datos.isnull().sum()
```

## Los estudiantes "Alicia" y "Carlos" ya no forman parte del grupo. Por lo tanto, elimnalos de la base de datos.

```
eliminar_alumnos = ['Alicia', 'Carlos']
datos = datos[~datos['Nombre'].isin(eliminar_alumnos)]
```

```
datos
```

## Aplica un filtro que seleccione solo a los estudiantes que fueron aprobados.

```
seleccion = datos[datos['Aprobado'] == 'True']
seleccion
```

 ## Al revisar las calificaciones de los estudiantes aprobados, notamos que algunas calificaciones eran incorrectas. Las estudiantes que obtuvieron una calificacin de 7.0, en realidad tenan un punto extra que no se contabiliz. Por lo tanto, reemplaza las calificaciones de 7.0 en la base de datos por 8.0. Consejo: busca el mtodo replace.

```
calificaciones = datos.replace(7.0, 8.0)
calificaciones
```
